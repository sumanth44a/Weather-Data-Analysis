# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])
df['Month']=df['Date/Time'].dt.month
df['Date']=df['Date/Time'].dt.day
df['Time']=df['Date/Time'].dt.time
```

```python
df.head()
```

```python
# Line Plot for Temp_C
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Temp_C'])
plt.xlabel('Date')
plt.ylabel('Temp_C (Celsius)')
plt.title('Temp_C Variation')
plt.show()
```

```python
sns.displot(data=df, x="Month",y="Rel Hum_%",kind='kde')
```

```python
# Scatter Plot for Temp_C vs. Wind Speed
plt.figure(figsize=(8, 6))
plt.bar(df['Temp_C'], df['Wind Speed_km/h'])
plt.xlabel('Temp_C (Celsius)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Temp_C vs. Wind Speed')
plt.show()
```

```python
# Box Plot for Temp_C Distribution
plt.figure(figsize=(8, 6))
sns.boxplot(x='Month', y='Temp_C', data=df)
plt.ylabel('Temp_C (Celsius)')
plt.title('Temp_C Distribution')

plt.show()
```

```python
month=df.Month
temp=df['Temp_C']
sns.jointplot(x=month,y=temp,kind='kde',cmap='Blues')
```

```python

plt.figure(figsize=(8, 6))
plt.bar(df['Month'], df['Temp_C'])
plt.xlabel('Month')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature (Celsius) over Time')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Month'], df['Dew Point Temp_C'])
plt.xlabel('month')
plt.ylabel('Dew Point Temperature (Celsius)')
plt.title('Dew Point Temperature (Celsius) over Time')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Wind Speed_km/h'], bins=20)
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.title('Wind Speed Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Month'], df['Rel Hum_%'])
plt.xlabel('month')
plt.ylabel('Relative Humidity (%)')
plt.title('Relative Humidity (%) over Time')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(x=df['Wind Speed_km/h'], y=df['Visibility_km'],  fill=True)
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Visibility (km)')
plt.title('2D Density Plot of Wind Speed and Visibility')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Month'], df['Wind Speed_km/h'])
plt.xlabel('month')
plt.ylabel('Wind Speed (km/h)')
plt.title('Wind Speed (km/h) over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Temp_C', data=df)
plt.xlabel('month')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature (Celsius) over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Wind Speed_km/h', data=df)
plt.xlabel('month')
plt.ylabel('Wind Speed (Km/h)')
plt.title('Wind Speed (Km/h) over Time')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(x='Temp_C', y='Rel Hum_%', data=df)
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Relative Humidity (%)')
plt.title('Temperature vs. Relative Humidity')
plt.xticks(df.Temp_C[::500])
plt.show()
```

```python
sns.pairplot(df[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Press_kPa']])
plt.show()
```

```python
correlation_matrix = df[['Temp_C', 'Dew Point Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Press_kPa']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Month'], df['Temp_C'], label='Temperature (Celsius)')
plt.bar(df['Month'], df['Dew Point Temp_C'], bottom=df['Temp_C'], label='Dew Point Temp (Celsius)')
plt.bar(df['Month'], df['Rel Hum_%'], bottom=df['Temp_C'] + df['Dew Point Temp_C'], label='Relative Humidity (%)')
plt.xlabel('Month')
plt.ylabel('Values')
plt.title('Temperature, Dew Point Temperature, and Relative Humidity over Time')
plt.legend()
plt.show()
```

```python
pies=pd.read_csv('Weather Data.csv')
```

```python
value_counts = pies['Weather'].value_counts(normalize=True)
values_to_delete = value_counts[(value_counts < 0.005)].index
pies = pies[~pies['Weather'].isin(values_to_delete)]
plt.figure(figsize=(8, 8))
weather_counts = pies['Weather'].value_counts()
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'plum', 'lightpink', 'lightgrey']
```

```python
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Weather Distribution (Custom Colors)')
plt.show()
```
